```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python

response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
```

```python
soup = BeautifulSoup(response, 'html.parser')
```

```python
soup
```

```python
soup.find('table').find_all('tr', class_='book')
```

```python
isbn = []
book = []
author = []
language = []
pages = []
topic = []
price = []
for books in soup.find('table').find_all('tr', class_= 'book'):
    isbn.append(books.find_all('td')[0].text)
    book.append(books.find_all('td')[1].text)
    author.append(books.find_all('td')[2].text)
    language.append(books.find_all('td')[3].text)
    pages.append(books.find_all('td')[4].text)
    topic.append(books.find_all('td')[5].text)
    price.append(books.find_all('td')[6].text)
```

```python
df = pd.DataFrame({
    'ISBN': isbn,
    'Book': book,
    'Author': author,
    'Language': language,
    'Pages': pages,
    'Topic': topic,
    'Price': price
})
```

```python
df['Pages'] = pd.to_numeric(df['Pages'], errors="coerce")
```

```python
df["Price"] = (
    df["Price"]
      .str.replace("$", "", regex=False)
      .astype(float)
)
```
