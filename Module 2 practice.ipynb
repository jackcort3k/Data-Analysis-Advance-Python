```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('csv files/companies.csv')

```

```python
df.head()
```

```python
plt.figure(figsize=(20, 7))
sns.countplot(x='state', data=df)
plt.show()
```

```python
df_sumofjobs = df.groupby('Industry', as_index=False).agg({'job_openings': 'sum'})

sns.barplot(x='job_openings', y='Industry', data=df_sumofjobs)
plt.show()
```

```python

sns.lineplot(x='founded', y='growth_percentage', data=df)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x= 'employees', y='estimated_revenues', data=df)
plt.show()
```

```python
df.columns
```

```python
df_companies = df.groupby('founded', as_index=False).agg({'company_name': 'count'})
plt.figure(figsize=(20, 10))
sns.histplot(y='company_name', x='founded', data=df_companies)
plt.show()
```
