```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('csv files/housing.csv')
```

```python
df.head(10)
```

```python
df.info()
```

```python
plt.figure(figsize=(20, 10))
sns.scatterplot(x='Landsize', y='Price', data=df, hue='Type')
plt.show()
```

```python
df_avgprice = df.groupby('YearBuilt', as_index=False).agg({'Price': 'mean'})
sns.lineplot(x='YearBuilt', y='Price', data=df_avgprice)
plt.show()
```

```python
df_avgprice = df.groupby('Regionname', as_index=False).agg({'Price': 'mean'})
plt.figure(figsize=(12, 7))
sns.barplot(x='Regionname', y='Price', data=df_avgprice)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.histplot(x='Price', data=df)
plt.show()
```
